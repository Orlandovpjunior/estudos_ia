```python
import cv2
import os 
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
# Escolher tarefa

tarefa = "segmentação" # "detecção" "pose"

if tarefa == "detecção":
    model = YOLO("yolov8n.pt")
elif tarefa == "segmentação":
    model = YOLO("yolov8n-seg.pt")
elif tarefa == "pose":
    model = YOLO("yolov8n-pose.pt")
```

```python
# Escolha da classe
# Dataset COCO
# 0 = person
# 1 = bicycle
# ...

classes_alvo= [0]
# classes_alvo= [0,1,2...]
```

```python
model.info()
```

```python
# Lista das imagens
extensoes_validas = (".jpg", ".jpeg", ".png", ".bmp", ".webp")
```

```python
arquivos_imagem = [
    arquivo for arquivo in os.listdir("./data/") if arquivo.lower().endswith(extensoes_validas)
]
```

```python
arquivos_imagem
```

```python
DATA_DIR = "./data/"
for img_path in arquivos_imagem:
    caminho = os.path.join(DATA_DIR, img_path)
    results = model(caminho, classes=classes_alvo)
    img_anotada = results[0].plot()
    
    for box in results[0].boxes:
        classe_id = int(box.cls[0])
        confianca = float(box.conf[0])
        nome_classe = model.names[classe_id]
        
        coordenadas=box.xyxy[0].tolist()
        
        print(f"Classe: {nome_classe}")
        print(f"Confiança: {confianca:.2f}")
        print(f"Coordenadas: {coordenadas}")
        print("-" * 30)
    
    imagem_original = cv2.imread(caminho)
    imagem_original_rgb = cv2.cvtColor(imagem_original, cv2.COLOR_BGR2RGB)
    imagem_anotada_rgb = cv2.cvtColor(img_anotada, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.imshow(imagem_original_rgb)
    plt.title("Imagem Original")
    plt.axis("off")
    
    plt.subplot(1,2,2)
    plt.imshow(imagem_anotada_rgb)
    plt.title("Imagem anotada")
    plt.axis("off")
    
    plt.show()
    
```
