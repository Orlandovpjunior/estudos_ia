```python
import numpy as np

# Configuração do perceptron para a porta and
pesos = np.array([0.5, 0.5])
bias = -0.7
```

```python
def funcao_ativacao(soma):
    return 1 if soma >= 0 else 0

def calcular_saida(entrada, pesos, bias):
    soma_ponderada = np.dot(entrada, pesos) + bias
    saida = 1 if soma_ponderada >= 0 else 0
    
    return saida, soma_ponderada 
```

```python
entradas = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

print("=== PERCEPTRON - PORTA LÓGICA AND ===")
print(f"Pesos: {pesos}")
print(f"Bias: {bias}")
```

```python
for entrada in entradas:
    saida, soma_ponderada = calcular_saida(entrada, pesos, bias)

    print(f"\nEntrada: {entrada}")
    print(f"\nCáclulo: ({pesos[0]} x {entrada[0]}) + ({pesos[1]} x {entrada[1]}) + ({bias}) = {soma_ponderada}")
    print(f"Função de ativação: {soma_ponderada} >= 0? {'Sim' if soma_ponderada >=0 else 'Não'} -> Saída: {saida}")
```
