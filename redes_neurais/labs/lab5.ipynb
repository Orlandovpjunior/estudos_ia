```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y =np.array([
    0,
    0,
    0,
    1
])

print("PERCEPTRON SIMPLES")

for i in range(len(X)):
    print(f'Entrada: {X[i]} -> Sáida esperada: {y[i]}')
```

```python
perceptron = Perceptron(max_iter=100,eta0=0.01, random_state=42,verbose=1)

perceptron.fit(X,y)

print("Resultados:")
print(f"N° de épocas: {perceptron.n_iter_}")
print(f"Pesos : {perceptron.coef_}")
print(f"Bias: {perceptron.intercept_}")
```

```python
previsoes = perceptron.predict(X)

print("Previsões")
print("Entrada | Saída Esperada | Previsão | Correto?")
print("==============================================")

for i in range(len(X)):
    correto = "sim" if previsoes[i] == y[i] else "não"

    print(f"    {X[i]}    |    {y[i]}    |    {previsoes[i]}    |     {correto}    ")
```
