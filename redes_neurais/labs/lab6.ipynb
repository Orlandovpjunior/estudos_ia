```python
import numpy as np
```

```python
class MLP_XOR:

    def __init__(self):

        # Camada 1
        self.W1 = np.array([
            [20.0,-20.0],
            [-20.0, 20.0]
        ])

        self.b1 = np.array([-10.0,-10.0])

        # Camada 2
        self.W2 = np.array([
            [20.0,20.0],
            [0.0,0.0]
        ])

        self.b2 = np.array([-30.0,0.0])

        # Camada de saída
        self.W3 = np.array([
            [20.0],
            [0.0]
        ])

        self.b3 = np.array([-10.0])
    
    def sigmoid(self,x):
        return 1/(1 + np.exp(-x))
    
    def forward(self,X):
        print(f"Entrada: {X}")

        z1 = np.dot(X,self.W1.T) + self.b1
        a1 = self.sigmoid(z1)
        print(f'\nCamada oculta 1:')
        print(f'z1 = {X}.W1.T + b1 = {z1}:')
        print(f'a1 = sigmoid(z1) = {a1}:')

        z2 = np.dot(a1,self.W2.T) + self.b2
        a2 = self.sigmoid(z2)
        print(f'\nCamada oculta 2:')
        print(f'z2 = a1.W2.T + b2 = {z2}:')
        print(f'a2 = sigmoid(z2) = {a2}:')

        z3 = np.dot(a2,self.W3) + self.b3
        a3 = self.sigmoid(z3)
        print(f'\nCamada oculta 2:')
        print(f'z3 = a2.W3. + b2 = {z3}:')
        print(f'a3 = sigmoid(z2) = {a3}:')

        output = 1 if a3 > 0.5 else 0
        print(f"Saída Final: {output}")

        return output, a3

```

```python
mlp = MLP_XOR()

entradas = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

esperado =np.array([
    0,
    1,
    1,
    0
])

resultados = []

for i, entrada in enumerate(entradas):
    print(f"Teste {i + 1}")
    saida, valor_sigmoid = mlp.forward(entrada)
    resultados.append((entrada,saida,valor_sigmoid[0]))

for i, (entrada, saida, sigmoid_val) in enumerate(resultados):
    correto = "sim" if saida == esperado[i] else "não"
    print(f'   {entrada}   |   {saida}   |   {sigmoid_val:.6f}   |   {esperado[i]}   |    {correto}')
```
