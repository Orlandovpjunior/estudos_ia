```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

## Carregar os dados

```python
df_cells = pd.read_csv('./dataset/synthetic_cell_data.csv')
```

```python
df_cells.info()
```

## EDA

```python
len(df_cells['cell_type'].unique())
```

```python
percentual_cell_type = df_cells.value_counts('cell_type') / len(df_cells) * 100
px.bar(percentual_cell_type, color=percentual_cell_type.index)
```

## Treinamento 2 componentes

```python
X = df_cells.copy()
```

```python
numeric_features = [i for i in df_cells.columns if i.startswith('gene_')]
categorical_features = ['cell_type']

preprocess = ColumnTransformer(
    transformers=[
        ('num',StandardScaler(),numeric_features),
        ('cat',OneHotEncoder(),categorical_features)
    ]
)

X_transformed = preprocess.fit_transform(X)
X_transformed
```

```python
df_results = pd.DataFrame()
```

```python
for perplexity in range(5, 51):
    tsne = TSNE(n_components=2,perplexity=perplexity, max_iter=250, init='random', random_state=51)
    tsne_results = tsne.fit_transform(X_transformed)

    temp_df = pd.DataFrame(tsne_results, columns=['Componente 1', 'Componente 2'])
    temp_df['Perplexity'] = perplexity
    temp_df['cell_type'] = df_cells['cell_type'].values
    df_results = pd.concat([df_results, temp_df], axis=0)
```

```python
df_results.head()
```

```python
df_results.reset_index(drop=True, inplace=True)
```

```python
df_results.head()
```

```python
px.scatter(df_results, x='Componente 1', y='Componente 2',color='cell_type')
```

## Treinamento 3 Componentes

```python
X2 = df_cells.copy()

numeric_features2 = [i for i in df_cells.columns if i.startswith('gene_')]
categorical_features2 = ['cell_type']

preprocess2 = ColumnTransformer(
    transformers=[
        ('num',StandardScaler(),numeric_features2),
        ('cat',OneHotEncoder(),categorical_features2)
    ]
)

X_transformed2 = preprocess2.fit_transform(X2)
X_transformed2
```

```python
df_results2 = pd.DataFrame()
```

```python
for perplexity2 in range(5,51):
    tsne2 = TSNE(n_components=3,perplexity=perplexity2,init='random',random_state=51,max_iter=250)
    tsne2_results = tsne2.fit_transform(X_transformed2)

    temp_df2 = pd.DataFrame(tsne2_results, columns=['Componente 1', 'Componente 2', 'Componente 3'])
    temp_df2['Perplexity'] = perplexity2
    temp_df2['cell_type'] = df_cells['cell_type'].values
    df_results2 = pd.concat([df_results2, temp_df2], axis=0)
```

```python
df_results2.head()
```

```python
px.scatter_3d(df_results2,x='Componente 1', y='Componente 2', z='Componente 3', color='cell_type')
```

## Anlise

A aplicao do algoritmo t-SNE demonstrou que os dados de expresso gnica so altamente informativos para a classificao celular. Observou-se a formao de agrupamentos claros e distintos para a maioria dos tipos, com destaque para o type_1 e type_3, que se posicionaram em regies isoladas do espao latente. Embora os grupos type_0 e type_4 apresentem maior proximidade, a visualizao em 3D revelou uma separao latente que confirma a distino entre todas as classes estudadas. Portanto, as 50 dimenses originais foram reduzidas com sucesso, preservando a estrutura local dos dados.
